```python
from PIL import Image
```

```python
import google.generativeai as genai
```

```python
defect_image= Image.open('defect.jpg')
```

```python
defect_image
```

```python
import os
gemini_api_key= os.getenv('google-api-key-2')
genai.configure(api_key=gemini_api_key)
model= genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt= "Identify the structural defect in the image and suggest how it can be fixed."
```

```python
response=model.generate_content([prompt, defect_image], 
                       generation_config={'temperature':0.7})
```

```python
print(response.text)
```
